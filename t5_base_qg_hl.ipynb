```python
from transformers import pipeline
import re

# Load the question generation model
qg_pipeline = pipeline("text2text-generation", model="valhalla/t5-base-e2e-qg")

# Sample context (replace with user input in your application)
context = """Recycling is the process of converting waste materials into new products to prevent unnecessary waste and conserve resources. Common recyclables include paper, glass, plastic, and metal. By recycling, we reduce the need to extract raw materials, save energy, and decrease pollution. It's a simple yet effective step individuals can take toward protecting the environment and promoting sustainability."""

def get_difficulty():
    """Prompt user for difficulty level and validate input."""
    while True:
        difficulty = input("Choose difficulty level (easy, medium, hard): ").lower()
        if difficulty in ["easy", "medium", "hard"]:
            return difficulty
        print("Invalid input. Please choose easy, medium, or hard.")

def classify_question(question):
    """Classify a question's difficulty based on keywords and structure."""
    question = question.lower()
    # Easy: Simple "what is" or short questions
    if re.match(r"what is|what are|which is", question) and len(question.split()) < 10:
        return "easy"
    # Medium: Questions about applications or processes
    elif re.match(r"how does|what.*applications|where.*used", question):
        return "medium"
    # Hard: "Why" or inference-based questions, or longer
    elif re.match(r"why|how.*work", question) or len(question.split()) > 15:
        return "hard"
    return "medium"  # Default to medium if unclear

def filter_questions(questions, target_difficulty):
    """Filter questions to match the target difficulty."""
    filtered = []
    for question in questions:
        difficulty = classify_question(question)
        if difficulty == target_difficulty:
            filtered.append(question)
    return filtered if filtered else questions  # Return all if no match

def generate_questions(context, difficulty):
    """Generate and filter questions based on difficulty."""
    input_text = "generate questions: " + context
    result = qg_pipeline(input_text, max_length=128, num_beams=4)

    # Extract questions (T5 model often returns questions separated by a delimiter like <sep>)
    generated_text = result[0]['generated_text']
    questions = [q.strip() for q in generated_text.split("<sep>") if q.strip()]

    # Filter questions by difficulty
    filtered_questions = filter_questions(questions, difficulty)

    return filtered_questions

def main():
    print("Welcome to the Question Generator!")
    difficulty = get_difficulty()
    print(f"You selected {difficulty} difficulty.")

    # Generate and display questions
    questions = generate_questions(context, difficulty)
    print("\n📋 Generated Questions:\n")
    if questions:
        for i, question in enumerate(questions, 1):
            print(f"{i}. {question}")
    else:
        print("No questions generated for the selected difficulty.")

if __name__ == "__main__":
    main()
```

```python
from transformers import pipeline
import re
import random

# Load the question generation model
qg_pipeline = pipeline("text2text-generation", model="valhalla/t5-base-e2e-qg", device=0)  # Using cuda:0

# Sample context
context = """The Internet of Things refers to the network of physical devices—like smartwatches, thermostats, and refrigerators—connected to the internet, collecting and sharing data. These devices can communicate with each other and be controlled remotely, making life more efficient and automated. IoT has applications in smart homes, healthcare, agriculture, and even cities, where it helps optimize energy use and improve safety."""

def get_difficulty():
    while True:
        difficulty = input("Choose difficulty level (easy, medium, hard): ").lower()
        if difficulty in ["easy", "medium", "hard"]:
            return difficulty
        print("Invalid input. Please choose easy, medium, or hard.")

def get_num_questions():
    while True:
        try:
            num = int(input("How many questions (2-10)? "))
            if 2 <= num <= 10:
                return num
            print("Please enter a number between 2 and 10.")
        except ValueError:
            print("Invalid input. Please enter a number.")

def classify_question(question):
    question = question.lower()
    if re.match(r"what is|what are|which is", question) and len(question.split()) < 10:
        return "easy"
    elif re.match(r"how does|what.*applications|where.*used", question):
        return "medium"
    elif re.match(r"why|how.*work", question) or len(question.split()) > 15:
        return "hard"
    return "medium"

def generate_unique_questions(context, difficulty, num_questions):
    input_text = "generate questions: " + context
    all_questions = set()  # Use a set to avoid duplicates
    attempts = 0
    max_attempts = 3  # Limit to avoid infinite loops

    while len(all_questions) < num_questions and attempts < max_attempts:
        # Use sampling for diversity
        result = qg_pipeline(
            input_text,
            max_length=64,  # Reduced to avoid overly long questions
            num_beams=5,
            num_return_sequences=5,
            do_sample=True,  # Enable sampling
            temperature=0.7,  # Control randomness
            top_k=50,  # Limit to top 50 tokens
        )

        for res in result:
            questions = [q.strip() for q in res['generated_text'].split("<sep>") if q.strip()]
            for q in questions:
                if classify_question(q) == difficulty:
                    all_questions.add(q)  # Add to set to ensure uniqueness
        attempts += 1

    # Convert to list and trim or pad if needed
    questions_list = list(all_questions)
    random.shuffle(questions_list)
    if len(questions_list) < num_questions:
        # Fill with any remaining questions if not enough unique ones
        extra_result = qg_pipeline(input_text, max_length=64, num_beams=5, num_return_sequences=num_questions - len(questions_list))
        for res in extra_result:
            extra_questions = [q.strip() for q in res['generated_text'].split("<sep>") if q.strip()]
            for q in extra_questions:
                if q not in questions_list and len(questions_list) < num_questions:
                    questions_list.append(q)
    return questions_list[:num_questions]

def main():
    print("Welcome to the Question Generator!")
    difficulty = get_difficulty()
    num_questions = get_num_questions()
    print(f"You selected {difficulty} difficulty and {num_questions} questions.")

    questions = generate_unique_questions(context, difficulty, num_questions)
    print("\n📋 Generated Questions:\n")
    if questions:
        for i, question in enumerate(questions, 1):
            print(f"{i}. {question}")
    else:
        print("No questions generated for the selected difficulty.")

if __name__ == "__main__":
    main()
```
